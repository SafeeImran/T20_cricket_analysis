```python
import pandas as pd
```

```python
df = pd.read_csv("asiacup.csv")
```

```python
print(df.head())
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df['result'] = df['result'].str.strip().str.title()
df['toss'] = df['toss'].str.strip().str.title()
df['selection'] = df['selection'].str.strip().str.title()
```

```python
numeric_cols = df.select_dtypes(include=['int64','float64']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
```

```python
#Feature Engineering
df['win_binary'] = df['result'].apply(lambda x: 1 if x == 'Win' else 0)
df['toss_binary'] = df['toss'].apply(lambda x: 1 if x == 'Win' else 0) 
df['is chasing'] = df['selection'].apply(lambda x: 1 if x == 'Bownling' else 0)
df['batting_margin'] = df['run_scored'] - df['wicket_lost']
df['bowling_effectiveness'] = df['wicket_taken'] / (df['given_extras'] + 1)  # +1 avoids div by 0
```

```python
df = df.drop_duplicates()
```
